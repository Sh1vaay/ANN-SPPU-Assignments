```python
import numpy as np
```

```python

def relu(x):
    return np.maximum(0, x)

# def forward(X, W1, W2):
#     z1 = np.dot(X, W1)
#     a1 = relu(z1)
#     z2 = np.dot(a1, W2)
#     return z2

# threshold = 0.5
# z2 = forward(X_test, W1, W2)
# predictions = (z2 > threshold).astype(int)


def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)


def forward(X, W1, W2):
    z1 = np.dot(X, W1)
    a1 = relu(z1)
    z2 = np.dot(a1, W2)
    y_hat = relu(z2)
    return y_hat, a1, z1

def compute_loss(y_hat, y):
    loss = -np.mean(y * np.log(y_hat))
    return loss



def backprop(X, y, y_hat, a1, z1, W1, W2):
    m = X.shape[0]
    dL_dz2 = (y_hat - y) / m
    dL_dW2 = np.dot(a1.T, dL_dz2)
    dL_da1 = np.dot(dL_dz2, W2.T)
    dL_dz1 = dL_da1 * (z1 > 0)
    dL_dW1 = np.dot(X.T, dL_dz1)
    return dL_dW1, dL_dW2
```

```python
input_size = 10
hidden_size = 100
output_size = 3
learning_rate = 0.01

W1 = np.random.randn(input_size, hidden_size)
W2 = np.random.randn(hidden_size, output_size)
# For example, if you have a dataset with 1000 samples, where each sample has 10 features and there are 3 classes, you could define X_train and y_train as follows:

X_train = np.random.randn(1000, 10)
y_train = np.random.randint(0, 3, size=(1000,))
# In this example, X_train is a matrix with shape (1000, 10) where each row represents a sample and each column represents a feature. y_train is a vector with shape (1000,) where each element represents the class label of the corresponding sample in X_train.
num_classes = 3
y_train_one_hot = np.eye(num_classes)[y_train]
print(y_train_one_hot)
for epoch in range(1000):
    print("iteration:",epoch)
    print('Input data:')
    print(X_train)
    y_hat, a1, z1 = forward(X_train, W1, W2)
    print('Predicted probabilities:')
    print(y_hat)
    loss = compute_loss(y_hat, y_train_one_hot)
    if epoch % 100 == 0:
        print(f'Epoch: {epoch}, Loss: {loss}')
    dL_dW1, dL_dW2 = backprop(X_train, y_train_one_hot, y_hat, a1, z1, W1, W2)
    W1 -= learning_rate * dL_dW1
    W2 -= learning_rate * dL_dW2
    predictions = np.argmax(y_hat, axis=1)
    print('Predicted class labels:')
    print(predictions)
```
