```python
def sigmoid(x,der=False):
    if der==True:
        return sigmoid(x)*(1-sigmoid(x))
    return 1/(1+np.exp(-x))

# Generate training data
X = np.array([[int(i) for i in list('{0:08b}'.format(x))] for x in range(256)])
y = np.array([x % 2 for x in range(256)]).reshape(-1, 1)
print(X)
print(y)


# Initialize weights
np.random.seed(1)
weights = 2 * np.random.random((8, 1)) - 1

# Train the neural network
for epoch in range(10000):
    # Forward propagation
    layer0 = X
    layer1 = sigmoid(np.dot(layer0, weights),der=False)

    # Calculate error
    layer1_error = y - layer1

    # Backpropagation
    layer1_delta = layer1_error * sigmoid(layer1,der=True)

    # Update weights
    weights += np.dot(layer0.T, layer1_delta)





# # Test classification
# print(classify(9)) # odd
# print(classify(2)) # even
```

```python

# Test the neural network
test_input = np.array([[int(i) for i in list('{0:08b}'.format(x))] for x in range(256)])

print("test input:",test_input)
test_output = sigmoid(np.dot(test_input, weights),der=False)
print("test output:",test_output)
# Evaluate the model
accuracy = np.mean(np.round(test_output) == y)
print("\nAccuracy: %.2f%%" % (accuracy*100))

```

```python
# Classify input
def classify(x):
    binary_input = np.array([int(i) for i in list('{0:08b}'.format(x))])
    print("binary input:",binary_input)
    output = sigmoid(np.dot(binary_input, weights),der=False)
    print("output",output)
    if round(output[0]) == 0:
        return "even"
    else:
        return "odd"
```
