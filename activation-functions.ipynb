# 1.Sigmoid Activation function 

### It is a function which is plotted as ‘S’ shaped graph.
### Equation : A = 1/(1 + e-x)
### Nature : Non-linear. Notice that X values lies between -2 to 2, Y values are very steep. This means, small changes in x would also bring about large changes in the value of Y.
### Value Range : 0 to 1
### Uses : Usually used in output layer of a binary classification, where result is either 0 or 1, as value for sigmoid function lies between 0 and 1 only so, result can be predicted easily to be 1 if value is greater than 0.5 and 0 otherwis

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))
def sig_der(x):
    return sigmoid(x)*(1- sigmoid(x))
x=np.linspace(10,-10,101)
out=sigmoid(x)
```

## Plotting Sigmoid Activation Function

```python
plt.xlabel("values")
plt.ylabel("sigmoid(values)")
plt.title("Activation Function:Sigmoid")
plt.plot(x,out,x,sig_der(x))
plt.title('Sigmoizd Activation Function & Derivative')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 2.Tanh Activation function 

### The activation that works almost always better than sigmoid function is Tanh function also knows as Tangent Hyperbolic function. It’s actually mathematically shifted version of the sigmoid function. Both are similar and can be derived from each other.
### Equation :-tanh(x)=(e^x – e^-x) / (e^x + e^-x)
### Deravative: 1 - tanh(x)**2

### Value Range :- -1 to +1
### Nature :- non-linear
### Uses :- Usually used in hidden layers of a neural network as it’s values lies between -1 to 1 hence the mean for the hidden layer comes out be 0 or very close to it, hence helps in centering the data by bringing mean close to 0. This makes learning for the next layer much easier.

```python

import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
      return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def der_tanh(x):
      return 1 - tanh(x) * tanh(x)

```

## Plotting Tanh Activation Function

```python

x_data=np.linspace(10,-10,101)
y_data = tanh(x_data)
dy_data = der_tanh(x_data)

plt.xlabel("values")
plt.ylabel("Tanh(values)")
plt.plot(x_data, y_data, x_data, dy_data)
plt.title("Activation Function:hyperbolic tangent & Derivative")
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 3.RELU Activation function 
### It Stands for Rectified linear unit. It is the most widely used activation function. Chiefly implemented in hidden layers of Neural network.
### Equation :- A(x) = max(0,x). It gives an output x if x is positive and 0 otherwise.
### Value Range :- [0, inf)
### Nature :- non-linear, which means we can easily backpropagate the errors and have multiple layers of neurons being activated by the ReLU function.
 

```python
import numpy as np
import matplotlib.pyplot as plt


def ReLU(x):
    data = [max(0,value) for value in x]
    return np.array(data, dtype=float)

def der_ReLU(x):
    data = [1 if value>0 else 0 for value in x]
    return np.array(data, dtype=float)
```

## Plotting ReLU Activation Function

```python
x_data = np.linspace(-10,10,101)
y_data = ReLU(x_data)
dy_data = der_ReLU(x_data)

plt.plot(x_data, y_data, x_data, dy_data)
plt.xlabel("values")
plt.ylabel("ReLU(values)")
plt.title('ReLU Activation Function & Derivative')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 4.leaky RELU Activation function 
### Leaky Rectified Linear Unit, or Leaky ReLU, is a type of activation function based on a ReLU, but it has a small slope for negative values instead of a flat slope. The slope coefficient is determined before training, i.e. it is not learnt during training. 
### Equation:lerelu(x) =  x if x>0 
###                 lerelu(x) = 0.01 * x if x<=0 
### Derivative:d/dx lerelu(x) =  1 if x>0
###                   d/dx lerelu(x) = 0.01 if x<=0 


```python
import numpy as np
import matplotlib.pyplot as plt

def lReLU(x):
    data = [max(0.01*value,value) for value in x]
    return np.array(data, dtype=float)

def der_lReLU(x):
    data = [1 if value>0 else 0.01 for value in x]
    return np.array(data, dtype=float)
```

## Plotting Leaky ReLU Activation Function

```python

x_data = np.linspace(-10,10,101)
y_data = lReLU(x_data)
dy_data = der_lReLU(x_data)


plt.plot(x_data, y_data, x_data, dy_data)
plt.xlabel("values")
plt.ylabel("LeakyReLU(values)")
plt.title('LeakyReLU Activation Function & Derivative')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 5.Soft Max Activation function
### The softmax function is also a type of sigmoid function but is handy when we are trying to handle multi- class classification problems.

### Nature :- non-linear
### Equation:e^x / sum(e^x)
### Output:- The softmax function is ideally used in the output layer of the classifier where we are actually trying to attain the probabilities to define the class of each input.


```python
import matplotlib.pyplot as plt
import numpy as np
from derivative import dxdt

def softmax(t):
    return np.exp(t) / np.sum(np.exp(t))
def der_softmax(t):
    return dxdt(softmax(t), t)
```

## Plotting Softmax Activation Function

```python
x_data=np.linspace(10,-10,80)
y_data = softmax(x_data)
dy_data= der_softmax(x_data)
```

```python
plt.plot(x_data, y_data, x_data, dy_data)
plt.xlabel("values")
plt.ylabel("softmax(values)")
plt.title('softmax Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 6.Linear Activation function

### Range : -inf to +inf
### Uses : Linear activation function is used at just one place i.e. output layer.

# 7.ELU Activation function

### It is the same as RELU for positive input, but for negative inputs, ELU smoothes slowly whereas RELU smoothes sharply.
### 

```python
def elu(x):
    ls=[]
    for i in x:
        if i<0:
            ls.append(0.5*(np.exp(i) -1))
        else:
            ls.append(i)
    return ls

def der_elur(x):
    return dxdt(elu(x), x)

x = np.linspace(-10, 10,101)
```

## Plotting ELU Activation Function

```python
plt.plot(x, elu(x))
plt.xlabel("values")

plt.ylabel("elu(values)")
plt.title('ELU Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

```python
plt.plot(x, der_elur(x))
plt.xlabel("values")
plt.ylabel("elu(values)")
plt.title('ELU Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 8.SoftPlus Activation function
### Softplus function is a smoothed form of the Relu activation function and its derivative is the sigmoid function. It also helps in overcoming the dying neuron problem.
### Equation:softplus(x) = log(1 + exp(x))
### Derivative:d/dx softplus(x) = 1 / (1 + exp(-x))

```python
def Softplus(x): 
    return np.log(1 + np.exp(-np.abs(x)))
def der_Softplus(x): 
    return (1/(1 + np.exp(-x)))

```

## Plotting SoftPlus Activation Function

```python
x = np.linspace(-10,10,101)
plt.xlabel("values") 
plt.ylabel("Softplus(x)") 
plt.plot(x, Softplus(x))
plt.title('SoftPlus Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```

# 9.Swish Activation function
### Swish is an activation function, , where  a learnable parameter. Nearly all implementations do not use the learnable parameter , in which case the activation function is  ("Swish-1").
### Equation: f ( x ) = x ⋅ sigmoid ( β x ) 

```python
def swish(x, B=1):
    return x/(1 + np.exp(-B*x))
def der_swish(x):
    
    return dxdt(swish(x), x)
```

## Plotting Swish Activation Function

```python
x = np.linspace(-10,10,101)
plt.xlabel("values") 
plt.ylabel("swish(x)") 
plt.plot(x, swish(x),x,der_swish(x))
plt.title('Swish Activation Function')
plt.legend(['f(x)','f"(x)'])
plt.grid()
plt.show()
```
